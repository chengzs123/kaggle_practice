```python
import numpy as np
import pandas as pd
```

```python
lgm_pth = './LGBM_reg_no_name.csv'
log_pth = './logistic_reg.csv'
xgb_pth = './XGB_clf.csv'
submission_pth = './data/sample_submission.csv'

lgm_pred = pd.read_csv(lgm_pth)
log_pred = pd.read_csv(log_pth)
xgb_pred = pd.read_csv(xgb_pth)
submission = pd.read_csv(submission_pth)
```

```python
lgm_pred['Survived_lgm'] = lgm_pred['Survived']
log_pred['Survived_log'] = log_pred['Survived']
xgb_pred['Survived_xgb'] = xgb_pred['Survived']
submission = pd.concat([submission, lgm_pred['Survived_lgm'], log_pred['Survived_log'], xgb_pred['Survived_xgb']], axis=1, sort=False)
```

```python
submission.head()
```

```python
submission['Survived'] = (submission[[col for col in submission.columns if col.startswith('Survived_')]].sum(axis=1) >= 2).astype(int)
```

```python
submission[['PassengerId', 'Survived']].to_csv('voted.csv', index=False)
```
