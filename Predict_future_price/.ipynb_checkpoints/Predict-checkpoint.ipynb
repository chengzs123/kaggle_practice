# Baseline

```python
#要求预测2015年11月的月销售量，baseline是将10月份的销售量当做11月的销售量
#训练数据是按照每天的销售量，而预测结果要求输出一个月的销售量，这里使用groupby和agg函数
#将训练集中第33个月即2015年的10月份的商品销售量统计起来
#groupby分组函数
train_subset = train_data[train_data['date_block_num'] == 33]
group = train_subset[['shop_id', 'item_id', 'item_cnt_day']].groupby(['shop_id', 'item_id'])
#查看分组
group.groups
#agg()聚合函数，参考https://www.jianshu.com/p/f37ac2a7b6dc，基本用法agg({列名：对该列的操作}),操作可以是sum，max，min
group_sum = group.agg({'item_cnt_day' : 'sum'}).reset_index()
# group_min = group.agg({'item_cnt_day' : 'min'}).reset_index()
# group_max = group.agg({'item_cnt_day' : 'max'}).reset_index()
group_sum = group_sum.rename(columns={'item_cnt_day' : 'item_cnt_month'})
print(group_sum.head())
# print(group_min.head())
# print(group_max.head())
```

```python
#预测生成submission
#预测要求，必须包含两列：ID和item_cnt_month，预测结果在[0, 20], 使用均方误差RMSE
#baseline的预测结果为1.16777
test = pd.merge(test_data, group_sum, on=['shop_id', 'item_id'], how='left')
print(test.head())
#fillna方法和clip方法，clip方法限制大小
#这里的fillna原因是在测试集中存在训练集中没有的商品，会出现Nan，这里用0代替
test['item_cnt_month'] = test['item_cnt_month'].fillna(0).clip(0, 20)
test = test[['ID', 'item_cnt_month']]
print(test.head())
# test.to_csv('submission.csv')
#取消掉前面的索引
submission = test.set_index('ID')
print(submission.head())
submission.to_csv('submission_baseline.csv')
```

```python
def downcast_dtypes(df):
    cols_float64 = [c for c in df if df[c].dtype == 'float64']
    cols_int64_32 = [c for c in df if df[c].dtype in ['int64', 'int32']]
    df[cols_float64] = df[cols_float64].astype(np.float32)
    df[cols_int64_32] = df[cols_int64_32].astype(np.int16)
    return df
trian_data = downcast_dtypes(train_data)
test_data = downcast_dtypes(test_data)
train_data.info()
```

```python
from itertools import product
import numpy as np
import time
t1 = time.time()
#matrix存储所有可能的匹配情况，每个商店与每件商品的组合
matrix = []
cols = ['date_block_num','shop_id','item_id']
for i in range(34):
#     print(product([i], sales.shop_id.unique(), sales.item_id.unique()))
    sales = train_data[train_data.date_block_num==i]
    #product函数返回列表中参数笛卡尔积构成的元组 https://blog.csdn.net/happyday_d/article/details/86005024
    matrix.append(np.array(list(product([i], sales.shop_id.unique(), sales.item_id.unique()))))

matrix = pd.DataFrame(np.vstack(matrix), columns=cols)
matrix['date_block_num'] = matrix['date_block_num'].astype(np.int8)
matrix['shop_id'] = matrix['shop_id'].astype(np.int8)
matrix['item_id'] = matrix['item_id'].astype(np.int16)
matrix.sort_values(cols,inplace=True)
print(matrix.head())
t2 = time.time()
print(t2 - t1)

#分组得到一个月内在同一个商店内同一件商品的销售情况,训练集
train_group = train_data.groupby(['item_id', 'shop_id', 'date_block_num']).agg({'item_cnt_day' : 'sum'})
# train_group = train_group.rename(columns={'item_cnt_day' : 'item_cnt_month'})
#df.columns获取所有的列值，这里修改了列名，修改列名的两种方法(回顾)
train_group.columns = ['item_cnt_month']
train_group.reset_index(inplace=True)
print(train_group.head())
# train_group.groups
```

3、历史信息的特征

```python
#特征延迟
def lag_feature(df, lags, col):
    tmp = df[['date_block_num', 'shop_id', 'item_id', col]]
    for i in lags:
        shifted = tmp.copy()
        shifted.columns = ['date_block_num', 'shop_id', 'item_id', col + '_lag_' + str(i)]
        shifted['date_block_num'] += i
        df = pd.merge(df, shifted, on=['date_block_num', 'shop_id', 'item_id'], how='left')
#         print(df.head())
    return df
```

```python
#1、(每一项商店-商品)月销售量历史特征(延迟1,2,3,6,12个月，下面的特征也是一样)，会出现Nan值，对应的就是若干个月前没有对应的商品和商店月销售情况
matrix = lag_feature(matrix, [1, 2, 3, 6, 12], 'item_cnt_month')
```

```python
#2、每个月（所有商品与商店）的平均销售量历史特征
group_avg = matrix.groupby(['date_block_num']).agg({'item_cnt_month' : 'mean'})
group_avg.columns = ['date_avg_item_cnt']
group_avg.reset_index(inplace=True)

matrix = pd.merge(matrix, group_avg, on=['date_block_num'], how='left')
matrix['date_avg_item_cnt'] = matrix['date_avg_item_cnt'].astype(np.float16)
matrix = lag_feature(matrix, [1, 2, 3, 6, 12], 'date_avg_item_cnt')
matrix.drop(columns=['date_avg_item_cnt'], inplace=True)
```

```python
#3、每件商品的月平均销量历史特征
group_avg = matrix.groupby(['date_block_num', 'item_id']).agg({'item_cnt_month' : 'mean'})
group_avg.columns = ['item_avg_item_cnt']
group_avg.reset_index(inplace=True)

matrix = pd.merge(matrix, group_avg, on=['date_block_num', 'item_id'], how='left')
matrix['item_avg_item_cnt'] = matrix['item_avg_item_cnt'].astype(np.float16)
matrix = lag_feature(matrix, [1, 2, 3, 6, 12], 'item_avg_item_cnt')
matrix.drop(columns=['item_avg_item_cnt'], inplace=True)
```

```python
#4、每个商店的月平均销量历史特征
#这里回顾的时候查看一下，做了groupby只显示一列，做了reset_index之后变成多列
group_avg = matrix.groupby(['date_block_num', 'shop_id']).agg({'item_cnt_month' : 'mean'})
group_avg.columns = ['shop_avg_item_cnt']
group_avg.reset_index(inplace=True)

matrix = pd.merge(matrix, group_avg, on=['date_block_num', 'shop_id'], how='left')
matrix['shop_avg_item_cnt'] = matrix['shop_avg_item_cnt'].astype(np.float16)
matrix = lag_feature(matrix, [1, 2, 3, 6, 12], 'shop_avg_item_cnt')
matrix.drop(columns=['shop_avg_item_cnt'], inplace=True)
```

```python
#5、每类商品的月平均销量历史特征
group_avg = matrix.groupby(['date_block_num', 'item_category_id']).agg({'item_cnt_month' : 'mean'})
group_avg.columns = ['item_category_avg_item_cnt']
group_avg.reset_index(inplace=True)

matrix = pd.merge(matrix, group_avg, on=['date_block_num', 'item_category_id'], how='left')
matrix['item_category_avg_item_cnt'] = matrix['item_category_avg_item_cnt'].astype(np.float16)
matrix = lag_feature(matrix, [1, 2, 3, 6, 12], 'item_category_avg_item_cnt')
matrix.drop(columns=['item_category_avg_item_cnt'], inplace=True)
```
